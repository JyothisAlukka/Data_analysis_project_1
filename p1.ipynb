```python
#1.find top 10 highest reveue generating products 
df = pd.read_sql("SELECT TOP 10 product_id, SUM(sell_price) AS sales FROM df_orders GROUP BY product_id ORDER BY sales DESC", conn)
print(df)
```

```python
#2. find top 5 highest selling products in each region
query = """
WITH cte AS (
    SELECT region, product_id, SUM(sell_price) AS sales
    FROM df_orders
    GROUP BY region, product_id
)
SELECT *
FROM (
    SELECT *,
           ROW_NUMBER() OVER(PARTITION BY region ORDER BY sales DESC) AS rn
    FROM cte
) A
WHERE rn <= 5
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#3. find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023
query = """
WITH cte AS (
    SELECT YEAR(order_date) AS order_year, MONTH(order_date) AS order_month,
           SUM(sell_price) AS sales
    FROM df_orders
    GROUP BY YEAR(order_date), MONTH(order_date)
)
SELECT order_month,
       SUM(CASE WHEN order_year = 2022 THEN sales ELSE 0 END) AS sales_2022,
       SUM(CASE WHEN order_year = 2023 THEN sales ELSE 0 END) AS sales_2023
FROM cte
GROUP BY order_month
ORDER BY order_month;
"""

df = pd.read_sql(query, engine)
print(df)
```

```python
#4. for each category which month had highest sales 
query = """
WITH cte AS (
    SELECT category, FORMAT(order_date, 'yyyyMM') AS order_year_month,
           SUM(sell_price) AS sales 
    FROM df_orders
    GROUP BY category, FORMAT(order_date, 'yyyyMM')
)
SELECT *
FROM (
    SELECT *,
           ROW_NUMBER() OVER(PARTITION BY category ORDER BY sales DESC) AS rn
    FROM cte
) a
WHERE rn = 1;
"""

df = pd.read_sql(query, engine)
print(df)
```

```python
#5. which sub category had highest growth by profit in 2023 compare to 2022
query = """
WITH cte AS (
    SELECT sub_category, YEAR(order_date) AS order_year,
           SUM(sell_price) AS sales
    FROM df_orders
    GROUP BY sub_category, YEAR(order_date)
)
, cte2 AS (
    SELECT sub_category,
           SUM(CASE WHEN order_year = 2022 THEN sales ELSE 0 END) AS sales_2022,
           SUM(CASE WHEN order_year = 2023 THEN sales ELSE 0 END) AS sales_2023
    FROM cte 
    GROUP BY sub_category
)
SELECT TOP 1 *,
       (sales_2023 - sales_2022) AS growth
FROM cte2
ORDER BY growth DESC;
"""

df = pd.read_sql(query, engine)
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sqlalchemy as sal

query = """
WITH cte AS (
    SELECT region, product_id, SUM(sell_price) AS sales
    FROM df_orders
    GROUP BY region, product_id
)
SELECT region, product_id, sales
FROM (
    SELECT *,
           ROW_NUMBER() OVER(PARTITION BY region ORDER BY sales DESC) AS rn
    FROM cte
) A
WHERE rn <= 5;
"""

df = pd.read_sql(query, engine)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="region", y="sales", hue="product_id", data=df)
plt.title("Top 5 Highest-Selling Products in Each Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.legend(title="Product ID", bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.show()
```

```python
query = """
WITH cte AS (
    SELECT YEAR(order_date) AS order_year, MONTH(order_date) AS order_month,
           SUM(sell_price) AS sales
    FROM df_orders
    GROUP BY YEAR(order_date), MONTH(order_date)
)
SELECT order_month,
       SUM(CASE WHEN order_year = 2022 THEN sales ELSE 0 END) AS sales_2022,
       SUM(CASE WHEN order_year = 2023 THEN sales ELSE 0 END) AS sales_2023
FROM cte
GROUP BY order_month
ORDER BY order_month;
"""

df = pd.read_sql(query, engine)

# Plot
df.plot(x="order_month", y=["sales_2022", "sales_2023"], kind="bar", figsize=(10, 6))
plt.title("Month-over-Month Sales Growth (2022 vs 2023)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.legend(["2022 Sales", "2023 Sales"])
plt.show()
```

```python
query = """
WITH cte AS (
    SELECT category, FORMAT(order_date, 'yyyyMM') AS order_year_month,
           SUM(sell_price) AS sales 
    FROM df_orders
    GROUP BY category, FORMAT(order_date, 'yyyyMM')
)
SELECT category, order_year_month, sales
FROM (
    SELECT *,
           ROW_NUMBER() OVER(PARTITION BY category ORDER BY sales DESC) AS rn
    FROM cte
) a
WHERE rn = 1;
"""

df = pd.read_sql(query, engine)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="category", y="sales", data=df)
plt.title("Month with Highest Sales for Each Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
query = """
WITH cte AS (
    SELECT sub_category, YEAR(order_date) AS order_year,
           SUM(sell_price) AS sales
    FROM df_orders
    GROUP BY sub_category, YEAR(order_date)
)
, cte2 AS (
    SELECT sub_category,
           SUM(CASE WHEN order_year = 2022 THEN sales ELSE 0 END) AS sales_2022,
           SUM(CASE WHEN order_year = 2023 THEN sales ELSE 0 END) AS sales_2023
    FROM cte 
    GROUP BY sub_category
)
SELECT sub_category, (sales_2023 - sales_2022) AS growth
FROM cte2
ORDER BY growth DESC;
"""

df = pd.read_sql(query, engine)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="sub_category", y="growth", data=df)
plt.title("Sub-Category with Highest Growth (2023 vs 2022)")
plt.xlabel("Sub-Category")
plt.ylabel("Sales Growth")
plt.xticks(rotation=45)
plt.show()
```
